```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Step, Time, Kin,Temp,Pot,ConsQty,UsedTime = np.loadtxt('../Outputs/NVT-1.ener', skiprows=1, unpack=True)
```

```python
plt.plot(Time, Pot)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs. Time in NVT')
plt.savefig(f'../Figures/Potential_Energy_vs_Time_NVT.pdf',format="pdf")
print('Plot saved: ../Figures/Potential_Energy_vs_Time_NVT.pdf')

plt.show()
```
