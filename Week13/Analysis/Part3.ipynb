```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Step, Time, Kin,Temp,Pot,ConsQty,UsedTime = np.loadtxt('../Outputs/NPT-1.ener', skiprows=1, unpack=True)
```

```python
plt.plot(Time, Pot)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs. Time in NVT')
plt.savefig(f'../Figures/Potential_Energy_vs_Time_NPT.pdf',format="pdf")
print('Plot saved: ../Figures/Potential_Energy_vs_Time_NPT.pdf')

plt.show()
```

```python
qmmm = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog')
metad1 = np.loadtxt('../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog')
metad2 = np.loadtxt('../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog')
```

```python
plt.plot(qmmm[:,0], qmmm[:,1], label='QMMM')
plt.plot(metad1[:,0], metad1[:,1], label='Matadynamic: frequency=1/100')
plt.plot(metad2[:,0], metad2[:,1], label='Matadynamic: frequency=1/20')
plt.xlabel('Time(fs)')
plt.ylabel('Reaction Coordinate($d_1 - d_2$)')
plt.legend(fontsize=8)
plt.title('Reaction Coordinate vs Time \n standard QM/MM & two Matadynamic runs with hill height=1.5 width=0.5',fontsize=14)
plt.savefig("../Figures/Reaction_Coordinate_vs_Time.pdf",format="pdf")
print("Plot saved: ../Figures/Reaction_Coordinate_vs_Time.pdf")

plt.show()
```
