```python
traj = mda.Universe('../Runs/Water/water_aimd_1ps_every0.5fs-pos-1.xyz')
traj

traj.dimensions = np.array([12.42, 12.42, 12.42, 90, 90, 90])
```

```python
box_size = 12.42
rdf = InterRDF(traj.select_atoms('name O'), traj.select_atoms('name O'), range=(0.0, box_size / 2), nbins=100)
rdf.run()
```

```python
plt.plot(rdf.bins[1:], rdf.rdf[1:])

plt.xlabel('Distance (Angstrom)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function for Oxygen in water')
plt.savefig(f'../Figures/Radial_Distribution_Function_for_Oxygen_water.pdf',format="pdf")
print('Plot saved: ../Figures/Radial_Distribution_Function_for_Oxygen_water.pdf')


plt.show()
```
